# Tema 7. La transformada Z

En este tema vamos a describir la transformada Z, una herramienta similar pero alternativa a la transformada de Fourier en tiempo discreto. Respecto a la transformada de Fourier, existen dos ventajas principales de la transformada Z:

<ol>
  <li> Mayor facilidad para estudiar las propiedades de los sistemas.
  <li> Hay señales que no tienen transformada de Fourier, pero sí transformada Z, lo que permite estudiar más situaciones como los sistemas caracterizados por ecuaciones en diferencias en sistemas inestables.
</ol>
