# Interconexión de sistemas
Muchos sistemas complicados están construidos mediante interconexión de sistemas elementales.

:::{prf:example} 
:class: dropdown
**Cadena musical**

Interconexión de: sintonizador de radio, reproductor de CD, pletinas, amplificador, altavoces.
:::
