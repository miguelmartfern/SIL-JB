# Propiedades básicas de los sistemas
\label{propiedades_sistemas}
Analizamos las propiedades más importantes que pueden tener los sistemas, tanto continuos como discretos\footnote{Aunque sólo se pongan las definiciones para sistemas continuos, para el caso discreto son totalmente equivalentes.}.

Antes de empezar a ver las propiedades de los sistemas, es conveniente indicar la **forma de proceder** para demostrar que un sistema tiene o no una propiedad\footnote{Para **demostrar** que un sistema posee o no una propiedad no es suficiente con enunciar la definición y sin demostrar cosa alguna, llegar a una determinada conclusión,}:

     - Si queremos **demostrar que un sistema posee una cierta propiedad**, debemos hacerlo **a partir de la definición** y debe cumplirse para todos los elementos que aparezcan en la definición (señales, instantes de tiempo, desplazamientos, constantes, ...).
	- Si queremos **demostrar que un sistema no posee una cierta propiedad**, es suficiente con que no se cumpla la definición para una determinada señal de entrada, o en un determinado instante de tiempo, o valor de las constantes involucradas, .... Por tanto, en este caso la demostración se puede realizar de dos formas alternativas:
	
	     - **A partir de la definición** como en el caso afirmativo, pero llegando a la conclusión de que en general no se cumple la definición.
		- **Mediante un contraejemplo** concreto que demuestre que se incumple la definición.
	

## Sistemas con y sin memoria

     - **Sistema sin memoria**: la **salida** en cada instante de tiempo ($t_0, n_0$) depende sólo de la **entrada** en ese mismo instante de tiempo ($t_0, n_0$).
     - **Sistema con memoria**: la **salida** en cada instante de tiempo depende de la **entrada** en un cierto instante de tiempo distinto (también puede depender del instante actual).

**Ejemplos**:

	- $y[n]=(2x[n]-x^2[n])^2$,\qquad  Sin memoria.
	- $y(t)=Rx(t)$,\qquad Sin memoria (resistencia).
	- $y(t)=x(t),\quad y[n]=x[n]$,\qquad Sin memoria (identidad).
	- $y[n]=x[n-1]$,\qquad Con memoria (retardo).
	- $y[n]=\sum\limits_{k=-\infty}^n x[k]$,\qquad Con memoria (acumulador o sumador).
	- $y(t)=\frac{1}{C}\int_{-\infty}^t x(\tau)d\tau$,\qquad Con memoria (condensador).

\vspace*{.5cm}
El concepto de memoria corresponde a cualquier mecanismo que permite almacenar (recordar) información de entradas pasadas (o futuras).

Por ejemplo, en el acumulador:
```{math}
y[n]=\sum\limits_{k=-\infty}^n x[k]=x[n]+\sum\limits_{k=-\infty}^{n-1} x[k]=x[n]+y[n-1].
```

     - En los sistemas físicos: memoria $\Rightarrow$ almacenamiento de energía.
	
		- Condensador: almacenamiento de carga eléctrica.
		- Automóvil: almacenamiento de energía cinética.
	
	- Sistemas digitales: memoria $\Rightarrow$ almacenamiento de valores en registros.

**Nota**: aunque el concepto de ``memoria'' sugiere recordar el pasado, no es necesario; también se puede referir al futuro. Esto puede no parecer real, pero puede ocurrir, por ejemplo, si la variable independiente es espacial (en imagen). En este caso el valor de salida en un pixel puede estar relacionado con el valor de entrada en los vecinos (``pasados'' y ``futuros'').
```{math}
y[n1,n2]=\sum\limits_{p=-N}^N\sum\limits_{q=-N}^N x[n1-p,n2-q].
```

## Causalidad

	- **Sistema causal**: hay varias definiciones equivalentes:
	
		- la **salida** en cualquier instante de tiempo depende sólo de valores de la **entrada** en el momento presente y en el pasado.
		```{math}
y(t_0) \xrightarrow{\text{depende de}} x(t),\ t\le t_0,\ \forall t_0.
```
		- Son sistemas **no anticipativos**: si dos entradas son idénticas hasta cierto instante, las salidas correspondientes deben ser también iguales hasta ese instante.
		```{math}
\text{Si}\ x_1(t)=x_2(t),\ \forall t\le t_0 \Rightarrow y_1(t)=y_2(t),\ \forall t\le t_0, \forall t_0.
```
	
	- **Sistema anticausal**:
	
		- la **salida** en cualquier instante de tiempo depende sólo de valores de la **entrada** en momentos futuros.
		```{math}
y(t_0) \xrightarrow{\text{depende de}} x(t),\ t>t_0,\ \forall t_0.
```
		- Son sistemas estrictamente anticipativos: si dos entradas son idénticas para tiempos posteriores a un cierto instante, las salidas correspondientes deben ser también iguales para tiempos posteriores a ese mismo instante.
		```{math}
\text{Si}\ x_1(t)=x_2(t),\ \forall t>t_0 \Rightarrow y_1(t)=y_2(t),\ \forall t>t_0, \forall t_0.
```
	
	- **Sistema no causal**: si no es causal ni anticausal.

**Ejemplos**:

	- Circuito RC,\qquad  Causal.\\[.75em]
	La salida, $v_c(t)$, depende sólo de valores presentes y pasados de $v_s(t)$.
	- Automóvil,\qquad Causal.\\[.75em]
	No anticipa acciones futuras del conductor.
	- $y[n]=-x[n+1]+x^2[n+2]$,\qquad Anticausal.\\[.75em]
	La salida para todo instante de tiempo depende sólo de la entrada en instantes posteriores.
	- $y(t)=x(t+1)$,\qquad Anticausal.\\[.75em]
	La salida en todo instante de tiempo depende sólo de la entrada en instantes futuros.
	- $y(t)=x(t)x(t+1)$,\qquad No causal.\\[.75em]
	La salida depende de la entrada en el instante presente pero también en el futuro.
	- Sistema sin memoria, \qquad Causal.\\[.75em]
	La salida depende únicamente de la entrada en el instante presente

## Invertibilidad: sistemas inversos
Hay varias definiciones equivalentes para un **sistema invertible**:

	\item[$-$] Un sistema es invertible si distintas entradas producen distintas salidas.
	```{math}
\text{Si}\ x_1(t)\neq x_2(t) \Rightarrow y_1(t)\neq y_2(t),\ \forall x_1(t), x_2(t); \forall t.
```
	- Un sistema es invertible cuando existe un sistema inverso, $T^{-1}$ tal que cuando la entrada al mismo es $y(t)$, su salida es $x(t)$.
	```{math}
\exists\ T^{-1}\ /\ \text{si}\ y(t)=T\{x(t)\} \Rightarrow x(t)=T^{-1}\{y(t)\},\ \forall t.
```
	- Un sistema es invertible cuando colocado en cascada con el sistema original, produce una salida $w(t)$ igual a la entrada $x(t)$ del primer sistema.
	
	```{figure} invertibilidad.png
```

	
	El sistema completo (en línea punteada en el dibujo) es el **sistema identidad**.

**Ejemplos**:

	- $y(t)=2x(t)$,\qquad Invertible.\\[.75em]
	Sistema inverso: $w(t)=\frac{1}{2}y(t)$.
	- $y[n]=\sum\limits_{k=-\infty}^n x[k]$,\qquad Invertible.\\[.75em]
	Sistema inverso: $w[n]=y[n]-y[n-1]$.
	- $y[n]=0$,\qquad No invertible.\\[.75em]
	Distintas entradas producen la misma salida.
	- $y(t)=x^2(t)$,\qquad No invertible.\\[.75em]
	No se puede determinar el signo de la entrada a partir de la salida.\\[.75em]
	$x_2(t)=-x_1(t)\ \Rightarrow\ y_2(t)=y_1(t)$.
	- Sistema de compresión sin pérdidas,\qquad Invertible.

## Estabilidad
Un **sistema** es **estable** si entradas acotadas producen salidas acotadas.
```{math}
\text{Si}\ |x(t)|<B_x,\ \forall t\ \Rightarrow\ |y(t)|<B_y,\ \forall t;\ B_x, B_y \in \real^+.
```

**Ejemplos**:

	- Sistemas físicos con disipación de energía: péndulo, circuito RC, automóvil con rozamiento.\qquad Estables.
	- Ecuación de cuenta bancaria,\qquad Inestable.\\[.75em]
	Salida crece indefinidamente aunque entrada esté acotada.
	- Acumulador: $y[n]=\sum\limits_{k=-\infty}^n x[k]$,\qquad Inestable.\\[.75em]
	Si, por ejemplo, $x[n]=u[n],\ |x[n]|\le 1,\forall t \Rightarrow y[n]=\sum\limits_{k=0}^n u[k]=(n+1)u[n]$. Crece sin límite.
	- $y(t)=tx(t)$,\qquad Inestable.\\[.75em]
	Si, por ejemplo, $x(t)=1,\ |x(t)|= 1,\forall t \Rightarrow y(t)=t$. Salida no acotada para entrada acotada.
	- $y(t)=e^{x(t)}$,\qquad Estable.\\[.75em]
	Si $|x(t)|<B_x,\ \forall t\ \Rightarrow\ |y(t)|<e^{B_x}=B_y,\ \forall t.$

\vspace*{.5cm}
Las dos últimas propiedades son fundamentales y las usaremos a lo largo de toda la asignatura: Invarianza en el tiempo y linealidad.

## Invarianza en el tiempo
Un sistema es invariante en el tiempo si sus propiedades se mantienen constantes con el tiempo.

En el lenguaje de las señales y sistemas, un **sistema** es **invariante en el tiempo** si un desplazamiento cualquiera en el tiempo de la señal de entrada produce ese mismo desplazamiento en la señal de salida.
```{math}
\text{Si}\ y(t)=T\{x(t)\}\ \Rightarrow\ y(t-t_0)=T\{x(t-t_0)\},\ \forall t_0.
```

Para que cumpla la propiedad debe verificarse para cualquier desplazamiento y cualquier señal de entrada.

Forma de demostrar que un sistema es invariante en el tiempo:

     - Sea $x_1(t)$ una entrada arbitraria $\Rightarrow$ $y_1(t)=T\{x_1(t)\}$.
	- Sea $x_2(t)=x_1(t-t_0)$ para un $t_0$ arbitrario $\Rightarrow$ $y_2(t)=T\{x_2(t)\}=T\{x_1(t-t_0)\}$.
	- El sistema es invariante en el tiempo si se cumple: $y_2(t)\stackrel{?}{=}y_1(t-t_0)$.

**Ejemplos**:

	- $y(t)=\sin[x(t)]$,\qquad Invariante en el tiempo.\\[.75em]
	Sea $x_1(t)$ una entrada arbitraria $\Rightarrow$ $y_1(t)=\sin[x_1(t)]$.\\[.5em]
	Sea $x_2(t)=x_1(t-t_0)$ para un $t_0$ arbitrario $\Rightarrow$ $y_2(t)=\sin[x_2(t)]=\sin[x_1(t-t_0)]$.\\[.5em]
	Por otro lado: $y_1(t-t_0)=\sin[x_1(t-t_0)]$.\\[.5em]
	Por tanto, se cumple que: $y_2(t)=y_1(t-t_0)$.
	- $y[n]=n x[n]$,\qquad Variante en el tiempo.\\[.75em]
	Se puede usar la demostración general, pero se puede comprobar fácilmente con un contraejemplo:\\[.5em]
	$x_1[n]=\delta[n]\ \Rightarrow\ y_1[n]=nx_1[n]=n\delta[n]=0.$\\[.5em]
	$x_2[n]=x_1[n-1]=\delta[n-1]\ \Rightarrow\ y_2[n]=nx_2[n]=n\delta[n-1]=\delta[n-1].$\\[.5em]
	Como $y_2[n]\neq y_1[n-1]$, el sistema no es invariante en el tiempo.
	- $y(t)=x(2t)$,\qquad Variante en el tiempo.\\[.75em]
	Usamos la demostración general\footnote{Sería fácil encontrar un contraejemplo, como $x_1(t)=u(t)$, y $t_0=1$.}:\\[.5em]
	Sea $x_1(t)$ una entrada arbitraria $\Rightarrow$ $y_1(t)=x_1(2t)$.\\[.5em]
	Sea $x_2(t)=x_1(t-t_0)$ para un $t_0$ arbitrario $\Rightarrow$ $y_2(t)=x_2(2t)=x_1(2t-t_0)$.\\[.5em]
	Por otro lado: $y_1(t-t_0)=x_1(2(t-t_0))=x_1(2t-2t_0)$.\\[.5em]
	Por tanto: $y_2(t)\neq y_1(t-t_0)$.
	

## Linealidad
Un **sistema** es **lineal** si posee la propiedad de superposición: si la señal de entrada al sistema es combinación lineal (superposición) de varias señales, la salida es la misma combinación lineal de las respuestas del sistema a cada una de las señales.
```{math}
\text{Si}\ x_3(t)=\alpha x_1(t)+\beta x_2(t)
```
```{math}
\Downarrow
```
```{math}
y_3(t)=\alpha T\{x_1(t)\}+\beta T\{x_2(t)\}=\alpha y_1(t)+\beta y_2(t),\ \forall x_1(t), x_2(t); \alpha, \beta\in\complex.
```

Esta propiedad se puede dividir en dos:

     - **Aditividad**: $\text{Si}\ x_3(t)=x_1(t)+x_2(t)\ \Rightarrow y_3(t)=y_1(t)+y_2(t).$
	- **Homogeneidad o escalamiento**: $\text{Si}\ x_2(t)=\alpha x_1(t)\ \Rightarrow y_2(t)=\alpha y_1(t).$

La propiedad de homogeneidad implica que en un **sistema lineal**, si la **entrada es nula**, la **salida** debe ser asimismo **nula**.
```{math}
\text{Si}\ \alpha=0:\ x(t)=0\ \Rightarrow\ y(t)=0.
```

Forma de demostrar que un sistema es lineal:

     - Sean $x_1(t)$ y $x_2(t)$ entradas arbitrarias. $y_1(t)=T\{x_1(t)\}$, $y_2(t)=T\{x_2(t)\}$.
	- Sea $x_3(t)=\alpha x_1(t)+\beta x_2(t)$, con constantes $\alpha$ y $\beta$ arbitrarias.
	- $y_3(t)=T\{x_3(t)\}=T\{\alpha x_1(t)+\beta x_2(t)\}$.
	- El sistema es lineal si se cumple: $y_3(t)\stackrel{?}{=}\alpha y_1(t)+\beta y_2(t)$.

**Ejemplos**:

	- $y(t)=tx(t)$,\qquad Lineal.\\[.75em]
	Sean $x_1(t)$ y $x_2(t)$ entradas arbitrarias. $y_1(t)=tx_1(t)$, $y_2(t)=tx_2(t)$.\\[.5em]
	Sea $x_3(t)=\alpha x_1(t)+\beta x_2(t)$ una combinación lineal cualquiera de las entradas anteriores.\\[.5em]
	$y_3(t)=tx_3(t)=t[\alpha x_1(t)+\beta x_2(t)]=\alpha tx_1(t)+\beta tx_2(t)$.\\[.5em]
	Se cumple que $y_3(t)=\alpha y_1(t)+\beta y_2(t)$.
	- $y(t)=x^2(t)$,\qquad No lineal.\\[.75em]
	Sean $x_1(t)$ y $x_2(t)$ entradas arbitrarias. $y_1(t)=x_1^2(t)$, $y_2(t)=x_2^2(t)$.\\[.5em]
	Sea $x_3(t)=\alpha x_1(t)+\beta x_2(t)$ una combinación lineal cualquiera de las entradas anteriores.\\[.5em]
	$y_3(t)=x_3^2(t)=(\alpha x_1(t)+\beta x_2(t))^2=\alpha^2 x_1^2(t)+\beta^2 x_2^2(t)+2\alpha\beta x_1(t)x_2(t)$.\\[.5em]
	Se ve claramente que $y_3(t)\neq\alpha y_1(t)+\beta y_2(t)$.\footnote{Se puede demostrar también mediante un contraejemplo, como por ejemplo: $x_1(t)=x_2(t)=\alpha=\beta=1$.}
	- $y[n]=\Re\{x[n]\}$,\qquad No lineal.\\[.75em]
	No cumple propiedad de homogeneidad\footnote{Las señales y constantes pueden ser complejas.}:\\[.5em]
	Poniendo una señal compleja cualquiera en forma cartesiana:\\ $x_1[n]=r[n]+js[n]\ \Rightarrow\ y_1[n]=r[n]$.\\[.5em]
	$x_2[n]=\alpha x_1[n]\ \Rightarrow\ y_2[n]=\Re\{\alpha r[n]+j\alpha s[n]\}$.\\[.5em]
	Por ejemplo, para $\alpha=j\ \Rightarrow\ y_2[n]=\Re\{jr[n]-s[n]\}=-s[n]$.\\[.5em]
	Se ve que $y_2[n]\neq \alpha y_1[n]=jr[n]$.

\newpage

## Ejemplos de propiedades

**Ejemplo 1**: $y(t)=e^{x(t)}$.

     - Memoria: La salida en cada instante de tiempo $t_0$ depende únicamente de la entrada en ese mismo instante: $y(t_0)=e^{x(t_0)},\ \forall t_0$.\quad $\rightarrow$\quad Sin memoria
	- Causalidad: Todo sistema sin memoria es causal.\quad $\rightarrow$\quad Causal.
	- Invertibilidad: $w(t)=\ln(y(t))=x(t)$. Se obtiene una señal idéntica a la entrada para cualquier instante de tiempo.\quad $\rightarrow$\quad Invertible.
	- Estabilidad: Si $|x(t)|<B_x,\ \forall t\ \Rightarrow\ |y(t)|<e^{B_x}=B_y,\ \forall t.$ \quad $\rightarrow$\quad Estable.
	- Invarianza en el tiempo: Sea $x_1(t)$ una entrada arbitraria $\Rightarrow$ $y_1(t)=e^{x_1(t)}$.\\[.5em]
	Sea $x_2(t)=x_1(t-t_0)$ para un $t_0$ arbitrario $\Rightarrow$ $y_2(t)=e^{x_2(t)}=e^{x_1(t-t_0)}$.\\[.5em]
	Por otro lado: $y_1(t-t_0)=e^{x_1(t-t_0)}$.\\[.5em]
	Por tanto, se cumple que: $y_2(t)=y_1(t-t_0)$.\quad $\rightarrow$\quad Invariante en el tiempo.
	- Linealidad: Sean $x_1(t)$ y $x_2(t)$ entradas arbitrarias. $y_1(t)=e^{x_1(t)}$, $y_2(t)=e^{x_2(t)}$.\\[.5em]
	Sea $x_3(t)=\alpha x_1(t)+\beta x_2(t)$ una combinación lineal cualquiera de las entradas anteriores.\\[.5em]
	$y_3(t)=e^{x_3(t)}=e^{\alpha x_1(t)}e^{\beta x_2(t)}$.\\[.5em]
	Como $y_3(t)\neq \alpha e^{x_1(t)}+\beta e^{x_2(t)}$.\quad $\rightarrow$\quad No lineal.

Por tanto el sistema es \{-M,+C,+I,+E,+IT,-L\}.

\vspace*{.5cm}
**Ejemplo 2**: $y[n]=x[n]x[n-1]$.

     - Memoria: La salida en cada instante de tiempo, $n_0$, depende de otro instante de tiempo distinto, $n_0-1$.\quad $\rightarrow$\quad Con memoria
	- Causalidad: La salida en cada instante de tiempo, $n_0$, depende de la entrada en ese mismo instante (presente) y en el anterior, $n_0-1$, (pasado).\quad $\rightarrow$\quad Causal.
	- Invertibilidad: Contraejemplo: Sean dos entradas distintas, $x_1[n]$ nula para las muestras pares y $x_2[n]$ nula para las muestras impares, pero $y_1[n]=y_2[n]=0$.\quad $\rightarrow$\quad No invertible\footnote{Otro contraejemplo sencillo es: $x_2[n]=-x_1[n]\ \Rightarrow\ y_2[n]=y_1[n]$.}.
	- Estabilidad: Si $|x[n]|<B_x,\ \forall n\ \Rightarrow\ |y[n]|<B_x^2=B_y,\ \forall n.$ \quad $\rightarrow$\quad Estable.
	- Invarianza en el tiempo: Sea $x_1[n]$ una entrada arbitraria $\Rightarrow$ $y_1[n]=x_1[n]x_1[n-1]$.\\[.5em]
	Sea $x_2[n]=x_1[n-n_0]$ para un $n_0$ arbitrario $\Rightarrow$ $y_2[n]=x_2[n]x_2[n-1]=x_1[n-n_0]x_1[n-n_0-1]$.\\[.5em]
	Por otro lado: $y_1[n-n_0]=x_1[n-n_0]x_1[n-n_0-1]$.\\[.5em]
	Por tanto, se cumple que: $y_2[n]=y_1[n-n_0]$.\quad $\rightarrow$\quad Invariante en el tiempo.
	- Linealidad: Sean $x_1[n]$ y $x_2[n]$ entradas arbitrarias.\\[.5em]
	$y_1[n]=x_1[n]x_1[n-1]$, $y_2[n]=x_2[n]x_2[n-1]$.\\[.5em]
	Sea $x_3[n]=\alpha x_1[n]+\beta x_2[n]$ una combinación lineal cualquiera de las entradas anteriores.\\[.5em]
	$y_3[n]=x_3[n]x_3[n-1]=(\alpha x_1[n]+\beta x_2[n])(\alpha x_1[n-1]+\beta x_2[n-1])$\\[.5em]
	$y_3[n]=\alpha^2 x_1[n]x_1[n-1]+\alpha\beta x_1[n]x_2[n-1]+\alpha\beta x_1[n-1]x_2[n]+\beta^2 x_2[n]x_2[n-1]$.\\[.5em]
	Como $y_3[n]\neq \alpha y_1[n]+\beta y_2[n]=\alpha x_1[n]x_1[n-1]+\beta x_2[n]x_2[n-1]$.\quad $\rightarrow$\quad No lineal.

Por tanto el sistema es \{+M,+C,-I,+E,+IT,-L\}.
