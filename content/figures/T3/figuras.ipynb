```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Configuracin de estilo
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'serif'

def setup_oppenheim_axes(ax, x_lim, y_lim, x_label, y_label):
    """Funcin auxiliar para dar estilo de libro a los ejes"""
    ax.grid(False)
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.set_xlim(x_lim)
    ax.set_ylim(y_lim)
    
    # Flechas
    ax.plot(x_lim[1], 0, ">k", markersize=6, clip_on=False)
    ax.plot(0, y_lim[1], "^k", markersize=6, clip_on=False)
    
    # Etiquetas
    # ### CAMBIO: Movemos la etiqueta horizontal ENCIMA del eje
    # Antes era: -y_lim[1]*0.15 (abajo).
    # Ahora es: +y_lim[1]*0.05 (un poco arriba) y va='bottom' (alineacin inferior)
    ax.text(x_lim[1], y_lim[1]*0.08, x_label, fontsize=16, ha='center', va='bottom')
    
    # Etiqueta vertical (se mantiene igual)
    ax.text(x_lim[1]*0.05, y_lim[1], y_label, fontsize=16, va='center')

# Parmetros del pulso
Tp = 1.0  
T_periods = [2.0, 4.0, 8.0] 

fig, axes = plt.subplots(3, 2, figsize=(16, 12))

# Ajuste de espacios (mantenemos el de la versin anterior)
plt.subplots_adjust(hspace=0.8, wspace=0.3)

# Ttulo General
fig.suptitle(r"De la Serie (SF) a la Transformada de Fourier (TF): $T_0 \to \infty$", fontsize=18, fontweight='bold', y=0.98)

for i, T in enumerate(T_periods):
    # --- DOMINIO DEL TIEMPO (Izquierda) ---
    ax_t = axes[i, 0]
    t = np.linspace(-T*1.5, T*1.5, 1000)
    
    pulse_train = np.where(np.abs(((t + T/2) % T) - T/2) <= Tp/2, 1, 0)
    
    if i == 2:
        pulse_train = np.where(np.abs(t) <= Tp/2, 1, 0)
        label_T = r"Aperidica ($T_0 \to \infty$)"
    else:
        label_T = r"Peridica $T_0 = {}$".format(int(T))
    
    ax_t.plot(t, pulse_train, color='#7E3258', linewidth=2)
    #setup_oppenheim_axes(ax_t, [-6, 6], [-0.2, 1.5], '$t$', '$x(t)$')
    
    # Ttulo subido
    ax_t.set_title(label_T, fontsize=16, y=1.15)
    
    # --- DOMINIO DE LA FRECUENCIA (Derecha) ---
    ax_f = axes[i, 1]
    
    w = np.linspace(-6*np.pi, 6*np.pi, 1000)
    envelope = Tp * np.sin(w*Tp/2) / (w*Tp/2)
    envelope[np.isnan(envelope)] = Tp 
    
    ax_f.plot(w, envelope, 'k--', alpha=0.3, linewidth=1)

    if i < 2:
        w0 = 2 * np.pi / T
        k_indices = np.arange(-20, 21) 
        w_k = k_indices * w0 
        ak = (Tp/T) * np.sin(w_k*Tp/2) / (w_k*Tp/2)
        ak[np.isnan(ak)] = Tp/T 
        visual_height = ak * T 
        
        ax_f.vlines(w_k, 0, visual_height, colors='#7E3258', linewidth=2)
        ax_f.plot(w_k, visual_height, 'o', color='#7E3258', markersize=4)
        if i == 0:
            title_freq = r"Discreto: $k\omega_0$"
        else:
            title_freq = r"Discreto: $k\omega_0$ (se densifica)"
        setup_oppenheim_axes(ax_f, [-15, 15], [-0.5, 1.5], r'$\omega$', r'$c_k$')

        # --- AQUI ESTA LA ETIQUETA OMEGA_0 ---
        # Dibujamos una flecha doble <-> entre 0 y w0
        # La ponemos en y = -0.3 para que no moleste a la grfica
        y_arrow = 0.3
        ax_f.annotate('', xy=(0, y_arrow), xytext=(w0, y_arrow),
                      arrowprops=dict(arrowstyle='<->', color='black', lw=1))
        # Ponemos el texto
        ax_f.text(w0/2, y_arrow + 0.4, r'$\omega_0$', ha='center', va='top', fontsize=16)
        setup_oppenheim_axes(ax_t, [-6, 6], [-0.2, 1.5], '$t$', r'$\tilde{x}(t)$')
    else:
        ax_f.plot(w, envelope, color='#7E3258', linewidth=2)
        ax_f.fill_between(w, 0, envelope, color='#7E3258', alpha=0.2)
        title_freq = r"Continuo: $X(\omega)$"
        setup_oppenheim_axes(ax_t, [-6, 6], [-0.2, 1.5], '$t$', '$x(t)$')
        setup_oppenheim_axes(ax_f, [-15, 15], [-0.5, 1.5], r'$\omega$', r'$X(\omega)$')
    
    
    # Ttulo subido
    ax_f.set_title(title_freq, fontsize=16, y=1.15)

plt.savefig('transicion_SF_a_TF.svg', dpi=300, bbox_inches='tight')
plt.savefig('transicion_SF_a_TF.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.animation import FuncAnimation, PillowWriter

# --- Configuracin de estilo ---
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'serif'

def setup_oppenheim_axes(ax, x_lim, y_lim, x_label_text, y_label_text):
    """Configura los ejes y devuelve las referencias a los textos."""
    ax.grid(False)
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.set_xlim(x_lim)
    ax.set_ylim(y_lim)
    
    # Flechas en los extremos
    ax.plot(x_lim[1], 0, ">k", markersize=6, clip_on=False)
    ax.plot(0, y_lim[1], "^k", markersize=6, clip_on=False)
    
    # Etiquetas
    # Eje X: Un poco por encima del eje
    lbl_x = ax.text(x_lim[1], y_lim[1]*0.04, x_label_text, fontsize=14, ha='center', va='bottom')
    
    # Eje Y: A la izquierda de la flecha, ligeramente bajado para no chocar con el ttulo
    # Usamos coordenadas relativas al lmite para posicionarlo bien
    lbl_y = ax.text(x_lim[1]*0.05, y_lim[1]*0.95, y_label_text, fontsize=14, va='center')
    
    return lbl_x, lbl_y

# --- Parmetros Iniciales ---
Tp = 1.0
color_main = '#7E3258'

# --- Configuracin de la figura ---
# Aadimos facecolor='white' para asegurar fondo blanco limpio
fig, (ax_t, ax_f) = plt.subplots(1, 2, figsize=(14, 5), facecolor='white')

# AJUSTE MANUAL AGRESIVO DE MRGENES
# left/right: Pegamos a los lados (0.05 y 0.98)
# bottom: Dejamos justo lo necesario para la etiqueta del eje X (0.05)
# top: Dejamos espacio para el ttulo principal (0.8)
plt.subplots_adjust(left=0.05, right=0.98, top=0.8, bottom=0.05, wspace=0.15)

# Ttulo General muy arriba
title_text = fig.suptitle(r"", fontsize=18, fontweight='bold', y=0.99)
Fourier_Text = fig.text(0.52, 0.5, '', fontsize=28, ha='center', va='center')

# --- INICIALIZACIN ---

# Limites de visualizacin fijos
XLIM_T = 12
XLIM_F = 15

# 1. Dominio del Tiempo
t = np.linspace(-15, 15, 2000)
line_t, = ax_t.plot([], [], color=color_main, linewidth=2)
# Eje Y un poco ms bajo (1.3) para que la flecha no pegue con el subttulo
lbl_tx, lbl_ty = setup_oppenheim_axes(ax_t, [-XLIM_T, XLIM_T], [-0.2, 1.2], '$t$', r'$\tilde{x}(t)$')

# 2. Dominio de la Frecuencia
w = np.linspace(-20, 20, 1000)
line_env, = ax_f.plot(w, Tp * np.sin(w*Tp/2) / (w*Tp/2), color=color_main, linewidth=2, alpha=0)
fill_env = ax_f.fill_between(w, 0, Tp * np.sin(w*Tp/2) / (w*Tp/2), color=color_main, alpha=0)

stem_lines = ax_f.vlines([], [], [], color=color_main, linewidth=2)
stem_markers, = ax_f.plot([], [], 'o', color=color_main, markersize=4)

lbl_fx, lbl_fy = setup_oppenheim_axes(ax_f, [-XLIM_F, XLIM_F], [-0.5, 1.25], r'$\omega$', r'$T_0 a_k$')

# --- UPDATE ---

def update(T):
    is_limit = (T >= 49)
    w0 = 2 * np.pi / T
    
    # Posicin vertical de los subttulos (y=1.15 ejes coordenados) para separarlos de la grfica
    title_y_pos = 1.04
    
    if not is_limit:
        title_text.set_text(r"Densificacin Espectral: $T_0 = {:.1f} s$, $\omega_0 = {:.1f} rad/s$".format(T, w0))
        ax_t.set_title(r"Seal Peridica", fontsize=14, y=title_y_pos)
        ax_f.set_title(r"Espectro Discreto", fontsize=14, y=title_y_pos)
        
        lbl_ty.set_text(r'$\tilde{x}(t)$')
        lbl_fy.set_text(r'$T_0 a_k$')
        
        Fourier_Text.set_text(r'$\overset{SF}{\longleftrightarrow}$')

        # --- CORRECCIN 1: Ocultar ticks si se salen del grfico ---
        # Dejamos un margen de seguridad (ej. 0.5) antes del lmite XLIM_T
        if T < (XLIM_T - 0.5):
            ax_t.set_xticks([-T, T])
            ax_t.set_xticklabels([r'$-T_0$', r'$T_0$'])
        else:
            # Si T es muy grande, borramos los ticks para que no deformen el eje
            ax_t.set_xticks([])

        # Ticks en Frecuencia
        if w0 > 0.4:
            ax_f.set_xticks([-w0, w0])
            ax_f.set_xticklabels([r'$-\omega_0$', r'$\omega_0$'])
        else:
            ax_f.set_xticks([])
            
    else:
        # CASO LMITE
        title_text.set_text(r"Lmite $T_0 \to \infty$: Transformada de Fourier")
        ax_t.set_title(r"Seal Aperidica", fontsize=14, y=title_y_pos)
        ax_f.set_title(r"Espectro Continuo", fontsize=14, y=title_y_pos)

        # fig.text(0.5, 0.50, r'$\overset{\mathcal{F}}{\longleftrightarrow}$', fontsize=28, ha='center', va='center')

        lbl_ty.set_text(r'$x(t)$')
        lbl_fy.set_text(r'$X(\omega)$')

        Fourier_Text.set_text(r'$\overset{\mathcal{F}}{\longleftrightarrow}$')

        ax_t.set_xticks([])
        ax_f.set_xticks([])

    # --- Grficas ---
    pulse_train = np.where(np.abs(((t + T/2) % T) - T/2) <= Tp/2, 1, 0)
    if is_limit:
         pulse_train = np.where(np.abs(t) <= Tp/2, 1, 0)
    line_t.set_data(t, pulse_train)

    if not is_limit:
        # Discreto
        line_env.set_alpha(0)
        fill_env.set_alpha(0)
        
        max_k = int(25 / w0) + 1
        k_indices = np.arange(-max_k, max_k + 1)
        w_k = k_indices * w0
        
        ak_scaled = Tp * np.sin(w_k*Tp/2) / (w_k*Tp/2)
        ak_scaled[np.isnan(ak_scaled)] = Tp
        
        segments = [ [(wk, 0), (wk, ak)] for wk, ak in zip(w_k, ak_scaled)]
        stem_lines.set_segments(segments)
        stem_lines.set_alpha(1)
        stem_markers.set_data(w_k, ak_scaled)
        stem_markers.set_alpha(1)
    else:
        # Continuo
        stem_lines.set_alpha(0)
        stem_markers.set_alpha(0)
        line_env.set_alpha(1)
        fill_env.set_alpha(0.3)

    return line_t, stem_lines, stem_markers, title_text, Fourier_Text, lbl_ty, lbl_fy, line_env, fill_env

# --- GENERAR ---
frames_T = np.concatenate([
    np.linspace(2, 6, 25),
    np.linspace(6, 50, 40),
    [50] * 40
])

anim = FuncAnimation(fig, update, frames=frames_T, interval=100, blit=False)

print("Generando 'densificacion_fourier.gif'...")
anim.save(
    'densificacion_fourier.gif', 
    writer=PillowWriter(fps=15), 
    dpi=200
)
print("Hecho!")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib as mpl

# ==========================================
# 1. CONFIGURACIN DE ESTILO (OPPENHEIM)
# ==========================================
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.unicode_minus'] = False

# Paleta
COLOR_SIG = '#7E3258'   # Granate
COLOR_PHASE = '#2ca02c' # Verde
COLOR_AXIS = 'black'

# ==========================================
# 2. PARMETROS
# ==========================================
FACTOR_ESPACIADO = 2.5
RADIO_PULSO = 1.6
WIDTH = RADIO_PULSO
distancia = FACTOR_ESPACIADO * WIDTH

# Definiciones fundamentales
T0 = distancia      
w0 = 2 * np.pi / T0 

# Rangos de visualizacin
RANGO_T = T0 * 1.8
K_MAX_SHOW = 8 
LIM_K = K_MAX_SHOW + 0.5 

# ==========================================
# 3. FUNCIONES AUXILIARES
# ==========================================
def tri_pulse(t, w=1.0):
    return np.maximum(0, 1 - np.abs(t)/w)

def sinc_sq_envelope_k(k_val, w0, w=1.0):
    omega = k_val * w0
    omega = np.where(omega == 0, 1e-10, omega)
    return (np.sin(omega*w/2) / (omega*w/2))**2

# --- FUNCIN SETUP_AXES CORREGIDA ---
def setup_axes(ax, xlim, ylim, xlabel, ylabel=None):
    """
    Configura ejes centrados con flechas.
    CORRECCIN: La etiqueta Y se coloca a la derecha de la flecha superior.
    """
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    
    # 1. Centrar espinas
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    
    # 2. Dibujar Flechas
    ax.plot(xlim[1], 0, ">k", clip_on=False, markersize=6)
    ax.plot(0, ylim[1], "^k", clip_on=False, markersize=6)
    
    # 3. Etiqueta Eje X (Abajo a la derecha)
    ax.text(xlim[1], ylim[1]*0.05, xlabel, fontsize=14, ha='center', va='bottom')
    
    # 4. Etiqueta Eje Y (A la DERECHA de la flecha vertical)
    if ylabel:
        # Calculamos un pequeo offset horizontal basado en el rango total
        offset_x = (xlim[1] - xlim[0]) * 0.05 
        # va='center' para que quede alineada con la punta
        ax.text(offset_x, ylim[1], ylabel, fontsize=14, ha='left', va='center')

# ==========================================
# 4. PREPARACIN DE LA FIGURA
# ==========================================
fig = plt.figure(figsize=(14, 7), facecolor='white')
gs = fig.add_gridspec(2, 2, width_ratios=[1.2, 1], height_ratios=[1, 1], wspace=0.2, hspace=0.35)

# --- EJES 1: TIEMPO (Izquierda) ---
ax_t = fig.add_subplot(gs[:, 0])
setup_axes(ax_t, [-RANGO_T, RANGO_T], [-0.2, 1.5], r'$t$', r'$\tilde{x}(t - t_d)$')

# Objetos Tiempo
t_vals = np.linspace(-RANGO_T, RANGO_T, 1000)
line_signal, = ax_t.plot([], [], color=COLOR_SIG, lw=2.5)
line_ghost, = ax_t.plot([], [], '--', color=COLOR_SIG, alpha=0.3, lw=1)

# Texto general con el valor numerico
text_val_td = ax_t.text(RANGO_T*0.65, 1.3, "", fontsize=14, bbox=dict(facecolor='white', edgecolor='none', alpha=0.8))

# *** NUEVO: Objetos para el Tick Mvil de td ***
# Un punto o pequea linea vertical sobre el eje
tick_marker, = ax_t.plot([], [], '|k', markersize=10, markeredgewidth=1.5) 
# La etiqueta "td" que se mueve con el tick
tick_label = ax_t.text(0, -0.15, r'$t_d$', fontsize=12, ha='center', color='black')


# --- EJES 2: MAGNITUD (Derecha Arriba) ---
ax_mag = fig.add_subplot(gs[0, 1])
setup_axes(ax_mag, [-LIM_K, LIM_K], [-0.1, 1.2], r'$k$', r'$|d_k|$')

# Ticks enteros
ax_mag.set_xticks(np.arange(-K_MAX_SHOW, K_MAX_SHOW+1, 2))

# Dibujo Esttico (Magnitud)
k_indices = np.arange(-K_MAX_SHOW, K_MAX_SHOW + 1)
mag_ck = sinc_sq_envelope_k(k_indices, w0, w=WIDTH)
k_cont = np.linspace(-K_MAX_SHOW, K_MAX_SHOW, 500)
env_vals = sinc_sq_envelope_k(k_cont, w0, w=WIDTH)

ax_mag.plot(k_cont, env_vals, color='gray', linestyle=':', alpha=0.5, lw=1)
ax_mag.vlines(k_indices, 0, mag_ck, color=COLOR_SIG, lw=2)
ax_mag.plot(k_indices, mag_ck, 'o', color=COLOR_SIG, markersize=5)

# --- EJES 3: FASE (Derecha Abajo) ---
ax_ph = fig.add_subplot(gs[1, 1])
setup_axes(ax_ph, [-LIM_K, LIM_K], [-3.8, 3.8], r'$k$', r'$\angle d_k=-k(\omega_0\,t_d)$')

ax_ph.set_yticks([-np.pi, 0, np.pi])
ax_ph.set_yticklabels([r'$-\pi$', '0', r'$\pi$'])
ax_ph.set_xticks(np.arange(-K_MAX_SHOW, K_MAX_SHOW+1, 2))

# Objetos Fase
stem_lines = ax_ph.vlines([], [], [], color=COLOR_PHASE, lw=1.5)
stem_dots, = ax_ph.plot([], [], 'o', color=COLOR_PHASE, markersize=5)
guide_line, = ax_ph.plot([], [], color='gray', linestyle='--', alpha=0.4, lw=1)

# ==========================================
# 5. ANIMACIN
# ==========================================
def init():
    line_signal.set_data([], [])
    line_ghost.set_data([], [])
    stem_dots.set_data([], [])
    guide_line.set_data([], [])
    stem_lines.set_segments([])
    text_val_td.set_text("")
    
    # Inicializar tick mvil
    tick_marker.set_data([], [])
    tick_label.set_text("")
    
    return line_signal, line_ghost, stem_dots, guide_line, stem_lines, text_val_td, tick_marker, tick_label

def update(frame):
    # frame 0->1
    td = frame * T0 
    
    # --- TIEMPO ---
    sig_y = np.zeros_like(t_vals)
    for k in range(-4, 5):
        center = k * T0 + td
        sig_y += tri_pulse(t_vals - center, w=WIDTH)
        
    sig_ghost = np.zeros_like(t_vals)
    for k in range(-4, 5):
        sig_ghost += tri_pulse(t_vals - k*T0, w=WIDTH)

    line_signal.set_data(t_vals, sig_y)
    line_ghost.set_data(t_vals, sig_ghost)
    # text_val_td.set_text(r"$t_d = {:.2f}$".format(td))
    
    # *** ACTUALIZAR POSICIN DEL TICK td ***
    tick_marker.set_data([td], [0]) # Marca en el eje (y=0)
    tick_label.set_position((td, -0.15)) # Etiqueta debajo
    tick_label.set_text(r'$t_d$')

    # --- FASE ---
    phase_vals = -k_indices * w0 * td
    phases_wrapped = (phase_vals + np.pi) % (2 * np.pi) - np.pi
    
    stem_dots.set_data(k_indices, phases_wrapped)
    
    segments = []
    for x, y in zip(k_indices, phases_wrapped):
        segments.append([(x, 0), (x, y)])
    stem_lines.set_segments(segments)
    
    # Aadir la frmula a la derecha (dentro del cuadro de la magnitud)
    ax_mag.text(0.95, 0.85, r'$d_k = c_k e^{-j k \omega_0 t_d}$', 
                transform=ax_mag.transAxes, fontsize=15, 
                ha='right', va='top', color='#444444',
                bbox=dict(facecolor='white', edgecolor='none', alpha=0.8))

    # Guia visual
    k_guide = np.linspace(-K_MAX_SHOW, K_MAX_SHOW, 100)
    guide_y = -k_guide * w0 * td
    guide_line.set_data(k_guide, guide_y)

    return line_signal, line_ghost, stem_dots, guide_line, stem_lines, text_val_td, tick_marker, tick_label

# ==========================================
# 6. GUARDAR
# ==========================================
total_frames = 60
ani = animation.FuncAnimation(fig, update, frames=np.linspace(0, 1, total_frames),
                              init_func=init, blit=True, interval=50)

print("Generando GIF final 'SF_desplazamiento.gif'...")
ani.save('SF_desplazamiento.gif', writer='pillow', fps=20, dpi=100)
print("Hecho!")

plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.animation import FuncAnimation, PillowWriter

# --- CONFIGURACIN DE ESTILO ---
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'serif'

# ==========================================
# PARMETROS
# ==========================================
RADIO_PULSO = 1.6  
RANGO_VISUAL_FIJO = 8.0 
WIDTH = RADIO_PULSO

# Arrays base
t = np.linspace(-RANGO_VISUAL_FIJO, RANGO_VISUAL_FIJO, 800)
w = np.linspace(-RANGO_VISUAL_FIJO, RANGO_VISUAL_FIJO, 800)

# --- COLORES ---
color_orig = '#7E3258'    # Granate (Original)
color_recon = '#1f77b4'   # Azul (Reconstruida)

# ==========================================
# FUNCIONES MATEMTICAS
# ==========================================
def tri_pulse(t, w=1.0):
    return np.maximum(0, 1 - np.abs(t)/w)

def sinc_sq_envelope(omega, w=1.0):
    omega = np.where(omega == 0, 1e-10, omega)
    val = (np.sin(omega*w/2) / (omega*w/2))**2
    return val

def reconstruction_sinc(axis, samples_pos, samples_val, step):
    recon = np.zeros_like(axis)
    for pos, val in zip(samples_pos, samples_val):
        recon += val * np.sinc((axis - pos) / step)
    return recon

def setup_axes(ax, xlim, ylim, xlabel, ylabel):
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_linewidth(1.2)
    ax.spines['left'].set_linewidth(1.2)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.plot(xlim[1], 0, ">k", markersize=6, clip_on=False)
    ax.plot(0, ylim[1], "^k", markersize=6, clip_on=False)
    ax.text(xlim[1], ylim[1]*0.04, xlabel, fontsize=16, ha='center', va='bottom')
    ax.text(xlim[1]*0.05, ylim[1], ylabel, fontsize=16, va='center')

def add_ellipses(ax, x_range, y_pos=0.4):
    ax.text(x_range[0]*0.95, y_pos, r'$\cdots$', fontsize=20, ha='center', color='black')
    ax.text(x_range[1]*0.95, y_pos, r'$\cdots$', fontsize=20, ha='center', color='black')

# =============================================================================
# FILA 1: MUESTREO EN FREC -> PERIODICIDAD EN TIEMPO (CON RECORTE EXACTO)
# =============================================================================
fig1, (ax_t1, ax_f1) = plt.subplots(1, 2, figsize=(14, 5), facecolor='white')
plt.subplots_adjust(wspace=0.3, top=0.80)

# --- PANEL IZQUIERDO (TIEMPO) ---
line_periodic_t1, = ax_t1.plot([], [], color=color_orig, linewidth=1.5, label='Suma Peridica')
line_window_t1, = ax_t1.plot([], [], color=color_recon, linestyle='--', linewidth=2.5, label='Central (Ventana)')
# Lneas verticales azules discontinuas
vlines_window_t1 = ax_t1.vlines([], [], [], color=color_recon, linestyle='--', linewidth=1.5)

lines_replicas_t1 = [] 

# --- PANEL DERECHO (FRECUENCIA) ---
line_recon_f1, = ax_f1.plot([], [], color=color_recon, linestyle='--', linewidth=2, label='Reconstruida $X_r(\omega)$')
ax_f1.plot(w, sinc_sq_envelope(w, w=WIDTH), color=color_orig, linewidth=2, label='Original $X(\omega)$')
stem_lines_f1 = ax_f1.vlines([], [], [], colors=color_orig, linewidth=2)
stem_markers_f1, = ax_f1.plot([], [], 'o', color=color_orig, markersize=4)

ax_f1.legend(loc='upper right', frameon=False, fontsize=10)
ax_t1.legend(loc='upper left', frameon=False, fontsize=10)

title_1 = fig1.suptitle("", fontsize=16, fontweight='bold', y=0.95)
cond_1 = fig1.text(0.5, 0.75, "", fontsize=14, ha='center', va='center')
fig1.text(0.5, 0.5, r'$\overset{\mathcal{F}}{\longleftrightarrow}$', fontsize=24, ha='center', va='center')

setup_axes(ax_t1, [-RANGO_VISUAL_FIJO, RANGO_VISUAL_FIJO], [-0.2, 1.8], r'$t$', r'$\tilde{x}(t)$')
setup_axes(ax_f1, [-RANGO_VISUAL_FIJO, RANGO_VISUAL_FIJO], [-0.2, 1.2], r'$\omega$', r'$X(\omega)$')
add_ellipses(ax_t1, [-RANGO_VISUAL_FIJO, RANGO_VISUAL_FIJO])

ax_t1.set_title(r"Tiempo: Enventanado del Periodo Central", fontsize=14, y=1.05)
ax_f1.set_title(r"Frecuencia: Muestreo e Interpolacin", fontsize=14, y=1.05)

def update_fila1(factor):
    global lines_replicas_t1
    T0 = factor * WIDTH
    w0 = 2 * np.pi / T0
    ALIASHING = factor < 2.0
    
    # Texto
    if ALIASHING:
        col = '#c00000'; txt = r"$T_0 < 2A$ (Solapamiento -> Aliasing Temporal)"
        tit = f"Factor {factor:.1f} (Aliasing)"
    else:
        col = 'black'; txt = r"$T_0 \geq 2A$ (Sin Solapamiento)"
        tit = f"Factor {factor:.1f} (Limpio)"
    title_1.set_text(tit); title_1.set_color(col)
    cond_1.set_text(txt); cond_1.set_color(col)
    
    # --- TIEMPO ---
    for l in lines_replicas_t1: l.remove()
    lines_replicas_t1 = []
    
    x_sum = np.zeros_like(t)
    k_max = int(RANGO_VISUAL_FIJO / T0) + 2
    for k in range(-k_max, k_max + 1):
        rep = tri_pulse(t - k*T0, w=WIDTH)
        x_sum += rep
        l, = ax_t1.plot(t, rep, color='gray', linestyle=':', lw=0.8, alpha=0.3)
        lines_replicas_t1.append(l)
    
    line_periodic_t1.set_data(t, x_sum)
    
    # Periodo Central
    limit = T0/2
    mask_window = np.abs(t) <= limit
    x_windowed = x_sum.copy()
    x_windowed[~mask_window] = np.nan
    line_window_t1.set_data(t, x_windowed)
    
    # --- CALCULO EXACTO DE LA ALTURA EN EL LIMITE ---
    if ALIASHING:
        if limit < RANGO_VISUAL_FIJO:
            # Calculamos el valor exacto de la suma en t = limit
            y_cut = 0
            for k in range(-k_max, k_max + 1):
                y_cut += tri_pulse(limit - k*T0, w=WIDTH)
            
            # Dibujamos las lneas solo hasta esa altura
            vlines_window_t1.set_segments([ [(-limit, 0), (-limit, y_cut)], [(limit, 0), (limit, y_cut)] ])
        else:
            vlines_window_t1.set_segments([])
    else:
        vlines_window_t1.set_segments([])

    ax_t1.set_xticks([x for x in [-T0, -T0/2, 0, T0/2, T0] if abs(x) <= RANGO_VISUAL_FIJO])
    ax_t1.set_xticklabels([r'$-T_0$', r'$-T_0/2$', '0', r'$T_0/2$', r'$T_0$'])
    
    # --- FRECUENCIA ---
    wk = np.arange(-(int(RANGO_VISUAL_FIJO/w0)+1), int(RANGO_VISUAL_FIJO/w0)+2) * w0
    samp_vals = sinc_sq_envelope(wk, w=WIDTH)
    X_rec = reconstruction_sinc(w, wk, samp_vals, step=w0)
    line_recon_f1.set_data(w, X_rec)
    
    stem_lines_f1.set_segments([[(x,0), (x,y)] for x,y in zip(wk, samp_vals)])
    stem_markers_f1.set_data(wk, samp_vals)
    ax_f1.set_xticks([-w0, 0, w0] if w0 < RANGO_VISUAL_FIJO else [0])
    ax_f1.set_xticklabels([r'$-\omega_0$', '0', r'$\omega_0$'] if w0 < RANGO_VISUAL_FIJO else ['0'])


# =============================================================================
# FILA 2: MUESTREO EN TIEMPO -> PERIODICIDAD EN FREC (CON RECORTE EXACTO)
# =============================================================================
fig2, (ax_t2, ax_f2) = plt.subplots(1, 2, figsize=(14, 5), facecolor='white')
plt.subplots_adjust(wspace=0.3, top=0.80)

# --- PANEL IZQUIERDO (TIEMPO) ---
line_recon_t2, = ax_t2.plot([], [], color=color_recon, linestyle='--', linewidth=2, label='Reconstruida $x_r(t)$')
ax_t2.plot(t, sinc_sq_envelope(t, w=WIDTH), color=color_orig, linewidth=2, label='Original $x(t)$')
stem_lines_t2 = ax_t2.vlines([], [], [], colors=color_orig, linewidth=2)
stem_markers_t2, = ax_t2.plot([], [], 'o', color=color_orig, markersize=4)

ax_t2.legend(loc='upper left', frameon=False, fontsize=10)

# --- PANEL DERECHO (FRECUENCIA) ---
line_periodic_f2, = ax_f2.plot([], [], color=color_orig, linewidth=1.5, label='Espectro Peridico')
line_window_f2, = ax_f2.plot([], [], color=color_recon, linestyle='--', linewidth=2.5, label='Filtro Paso Bajo')
# Lneas verticales azules discontinuas
vlines_window_f2 = ax_f2.vlines([], [], [], color=color_recon, linestyle='--', linewidth=1.5)

lines_replicas_f2 = []

ax_f2.legend(loc='upper right', frameon=False, fontsize=10)

title_2 = fig2.suptitle("", fontsize=16, fontweight='bold', y=0.95)
cond_2 = fig2.text(0.5, 0.75, "", fontsize=14, ha='center', va='center')
fig2.text(0.5, 0.5, r'$\overset{\mathcal{F}}{\longleftrightarrow}$', fontsize=24, ha='center', va='center')

setup_axes(ax_t2, [-RANGO_VISUAL_FIJO, RANGO_VISUAL_FIJO], [-0.2, 1.2], r'$t$', r'$x(t)$')
setup_axes(ax_f2, [-RANGO_VISUAL_FIJO, RANGO_VISUAL_FIJO], [-0.2, 1.8], r'$\omega$', r'$X_p(\omega)$')
add_ellipses(ax_f2, [-RANGO_VISUAL_FIJO, RANGO_VISUAL_FIJO])

ax_t2.set_title(r"Tiempo: Muestreo e Interpolacin", fontsize=14, y=1.05)
ax_f2.set_title(r"Frecuencia: Filtro Paso Bajo (Recorte)", fontsize=14, y=1.05)

def update_fila2(factor):
    global lines_replicas_f2
    ws = factor * WIDTH
    Ts = 2 * np.pi / ws
    ALIASHING = factor < 2.0
    
    if ALIASHING:
        col = '#c00000'; txt = r"$\omega_s < 2B$ (Solapamiento -> Aliasing en Frecuencia)"
        tit = f"Factor {factor:.1f} (Aliasing)"
    else:
        col = 'black'; txt = r"$\omega_s \geq 2B$ (Reconstruccin Perfecta)"
        tit = f"Factor {factor:.1f} (Limpio)"
    title_2.set_text(tit); title_2.set_color(col)
    cond_2.set_text(txt); cond_2.set_color(col)
    
    # --- TIEMPO ---
    n_max = int(RANGO_VISUAL_FIJO / Ts) + 1
    tn = np.arange(-n_max, n_max + 1) * Ts
    vals = sinc_sq_envelope(tn, w=WIDTH)
    x_rec = reconstruction_sinc(t, tn, vals, step=Ts)
    line_recon_t2.set_data(t, x_rec)
    
    stem_lines_t2.set_segments([[(x,0), (x,y)] for x,y in zip(tn, vals)])
    stem_markers_t2.set_data(tn, vals)
    ax_t2.set_xticks([x for x in [-Ts, 0, Ts] if abs(x) <= RANGO_VISUAL_FIJO])
    ax_t2.set_xticklabels([r'$-T_s$', '0', r'$T_s$'])
    
    # --- FRECUENCIA ---
    for l in lines_replicas_f2: l.remove()
    lines_replicas_f2 = []
    
    X_sum = np.zeros_like(w)
    k_max = int(RANGO_VISUAL_FIJO / ws) + 2
    for k in range(-k_max, k_max + 1):
        rep = tri_pulse(w - k*ws, w=WIDTH)
        X_sum += rep
        l, = ax_f2.plot(w, rep, color='gray', linestyle=':', lw=0.8, alpha=0.3)
        lines_replicas_f2.append(l)
    
    line_periodic_f2.set_data(w, X_sum)
    
    # Filtro Paso Bajo
    limit = ws/2
    mask_filter = np.abs(w) <= limit
    X_filtered = X_sum.copy()
    X_filtered[~mask_filter] = np.nan
    line_window_f2.set_data(w, X_filtered)
    
    # --- CALCULO EXACTO DE LA ALTURA EN EL LIMITE ---
    if ALIASHING:
        if limit < RANGO_VISUAL_FIJO:
            # Calculamos el valor exacto de la suma en w = limit
            y_cut = 0
            for k in range(-k_max, k_max + 1):
                y_cut += tri_pulse(limit - k*ws, w=WIDTH)
                
            vlines_window_f2.set_segments([ [(-limit, 0), (-limit, y_cut)], [(limit, 0), (limit, y_cut)] ])
        else:
            vlines_window_f2.set_segments([])
    else:
        vlines_window_f2.set_segments([])

    ax_f2.set_xticks([x for x in [-ws, -ws/2, 0, ws/2, ws] if abs(x) <= RANGO_VISUAL_FIJO])
    ax_f2.set_xticklabels([r'$-\omega_s$', r'$-\frac{\omega_s}{2}$', '0', r'$\frac{\omega_s}{2}$', r'$\omega_s$'])


# ==========================================
# GENERACIN DE GIFS
# ==========================================
frames_factor = np.arange(3.0, 0.8, -.05)

print("Generando Fila 1 (animacion_reconstruccion_frecuencia.gif)...")
anim1 = FuncAnimation(fig1, update_fila1, frames=frames_factor, interval=100, blit=False)
anim1.save('animacion_reconstruccion_frecuencia.gif', writer=PillowWriter(fps=5), dpi=200)

print("Generando Fila 2 (animacion_reconstruccion_tiempo.gif)...")
anim2 = FuncAnimation(fig2, update_fila2, frames=frames_factor, interval=100, blit=False)
anim2.save('animacion_reconstruccion_tiempo.gif', writer=PillowWriter(fps=5), dpi=200)

print("Hecho!")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.animation import FuncAnimation, PillowWriter

# --- CONFIGURACIN DE ESTILO ---
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'serif'

# ==========================================
# PARMETROS
# ==========================================
RANGO_VISUAL = 6.0
puntos = 1000

t = np.linspace(-RANGO_VISUAL, RANGO_VISUAL, puntos)
w = np.linspace(-RANGO_VISUAL, RANGO_VISUAL, puntos)

color_time = '#7E3258'    # Granate
color_freq = '#7E3258'    # Granate

# ==========================================
# FUNCIONES
# ==========================================
def gaussian_time(t, sigma):
    return np.exp(-t**2 / (2 * sigma**2))

def gaussian_freq(omega, sigma):
    # Normalizamos altura a 1
    return np.exp(-omega**2 * sigma**2 / 2)

def setup_axes(ax, xlim, ylim, xlabel, ylabel):
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_linewidth(1.2)
    ax.spines['left'].set_linewidth(1.2)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.plot(xlim[1], 0, ">k", markersize=6, clip_on=False)
    ax.plot(0, ylim[1], "^k", markersize=6, clip_on=False)
    ax.text(xlim[1], 0.1, xlabel, fontsize=18, ha='center', va='top')
    ax.text(0.2, ylim[1]*0.95, ylabel, fontsize=18, va='center')
    ax.set_xticks([])
    ax.set_yticks([])

# ==========================================
# CONFIGURACIN FIGURA
# ==========================================
fig, (ax_t, ax_f) = plt.subplots(1, 2, figsize=(14, 6), facecolor='white')
plt.subplots_adjust(wspace=0.2, top=0.85, bottom=0.15)

# --- PANEL IZQUIERDO (TIEMPO) ---
line_t, = ax_t.plot([], [], color=color_time, linewidth=2)
setup_axes(ax_t, [-RANGO_VISUAL, RANGO_VISUAL], [-0.1, 1.3], r'$t$', r'$x(t)$')
ax_t.set_title(r"Dominio del Tiempo", fontsize=16, y=1.05)

arrow_t = ax_t.annotate('', xy=(0,0), xytext=(0,0), 
                        arrowprops=dict(arrowstyle='<->', color='black', lw=1.5))
label_t = ax_t.text(0, 0, r"$\sigma$", fontsize=14, ha='center', fontweight='bold')

# --- PANEL DERECHO (FRECUENCIA) ---
line_f, = ax_f.plot([], [], color=color_freq, linewidth=2)
setup_axes(ax_f, [-RANGO_VISUAL, RANGO_VISUAL], [-0.1, 1.3], r'$\omega$', r'$X(\omega)$')
ax_f.set_title(r"Dominio de la Frecuencia", fontsize=16, y=1.05)

arrow_f = ax_f.annotate('', xy=(0,0), xytext=(0,0), 
                        arrowprops=dict(arrowstyle='<->', color='black', lw=1.5))
label_f = ax_f.text(0, 0, r"$1/\sigma$", fontsize=14, ha='center', fontweight='bold')

fig.text(0.5, 0.5, r'$\overset{\mathcal{F}}{\longleftrightarrow}$', fontsize=30, ha='center', va='center')
fig.suptitle("Reciprocidad Tiempo - Frecuencia (Escala Geomtrica)", fontsize=20, fontweight='bold')

# ==========================================
# BUCLE DE ANIMACIN
# ==========================================
def update(sigma):
    # 1. Calcular y actualizar lneas
    y_t = gaussian_time(t, sigma)
    y_f = gaussian_freq(w, sigma)
    line_t.set_data(t, y_t)
    line_f.set_data(w, y_f)
    
    # 2. Actualizar Flechas y Etiquetas
    h_level = 0.606 
    
    # Tiempo
    arrow_t.xy = (-sigma, h_level)
    arrow_t.set_position((sigma, h_level))
    label_t.set_position((0, h_level + 0.1))
    label_t.set_text(rf"$\sigma = {sigma:.2f}$")
    
    # Frecuencia
    inv_sigma = 1.0 / sigma
    arrow_f.xy = (-inv_sigma, h_level)
    arrow_f.set_position((inv_sigma, h_level))
    label_f.set_position((0, h_level + 0.1))
    label_f.set_text(rf"$1/\sigma = {inv_sigma:.2f}$")

# ==========================================
# GENERAR CON ESCALA GEOMTRICA
# ==========================================
# Usamos geomspace en lugar de linspace.
# Esto hace que los pasos sean multiplicativos en lugar de aditivos.
# Visualmente: el cambio relativo es constante.
sigmas = np.geomspace(0.25, 4.0, 80) # Empiezo un poco ms bajo (0.4) para notar ms el efecto
sigmas = np.concatenate([sigmas, sigmas[::-1]])

print("Generando animacin con paso geomtrico...")
anim = FuncAnimation(fig, update, frames=sigmas, interval=50, blit=False)
anim.save('reciprocidad_TF_gauss.gif', writer=PillowWriter(fps=20), dpi=300)
print("Hecho! Archivo guardado: reciprocidad_TF.gif")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.animation import FuncAnimation, PillowWriter

# --- CONFIGURACIN DE ESTILO ---
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'serif'

# ==========================================
# PARMETROS
# ==========================================
LIMITE_T = 4.0   # Rango para el tiempo (ej: de -4 a 4)
LIMITE_W = 2*np.pi*LIMITE_T  # Rango para frecuencia (ej: de -20 a 20)
puntos = 1000

t = np.linspace(-LIMITE_T, LIMITE_T, puntos)
w = np.linspace(-LIMITE_W, LIMITE_W, puntos)

color_time = '#7E3258'    # Granate
color_freq = '#7E3258'    # Granate

# ==========================================
# FUNCIONES
# ==========================================
def triangular_pulse(t, tau):
    return np.maximum(0, 1 - np.abs(t) / tau)

def sinc_squared_freq(omega, tau):
    # La TF terica tiene su primer nulo en 2*pi / tau
    # Usamos np.sinc(x) que es sin(pi*x)/(pi*x)
    # Queremos que sea cero cuando omega = 2pi/tau
    # Argumento np.sinc debe ser 2 -> x = omega*tau / 2pi NO,
    # np.sinc(1) = 0. Entonces x debe ser 1 cuando omega = 2pi/tau.
    # x = omega * tau / (2*pi)
    val = (np.sinc(omega * tau / (2 * np.pi)))**2
    return val

def setup_axes(ax, xlim, ylim, xlabel, ylabel):
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_linewidth(1.2)
    ax.spines['left'].set_linewidth(1.2)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.plot(xlim[1], 0, ">k", markersize=6, clip_on=False)
    ax.plot(0, ylim[1], "^k", markersize=6, clip_on=False)
    ax.text(xlim[1], 0.1, xlabel, fontsize=18, ha='center', va='top')
    ax.text(0.3, ylim[1]*0.95, ylabel, fontsize=18, va='center')
    ax.set_xticks([])
    ax.set_yticks([])

# ==========================================
# CONFIGURACIN FIGURA
# ==========================================
fig, (ax_t, ax_f) = plt.subplots(1, 2, figsize=(14, 6), facecolor='white')
plt.subplots_adjust(wspace=0.2, top=0.85, bottom=0.15)

# --- TIEMPO ---
line_t, = ax_t.plot([], [], color=color_time, linewidth=2.5)
setup_axes(ax_t, [-LIMITE_T, LIMITE_T], [-0.1, 1.2], r'$t$', r'$x(t)$')
# ax_t.set_title(r"Tiempo: Pulso de ancho $\tau$", fontsize=16, y=1.05)

arrow_t = ax_t.annotate('', xy=(0,0), xytext=(0,0), 
                        arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
label_t = ax_t.text(0, 0, '', fontsize=14, ha='center', fontweight='bold')

# --- FRECUENCIA ---
line_f, = ax_f.plot([], [], color=color_freq, linewidth=2)
setup_axes(ax_f, [-LIMITE_W, LIMITE_W], [-0.1, 1.2], r'$\omega$', r'$X(\omega)\propto\text{sinc}^2\left( \frac{\omega \tau}{2\pi} \right)$')
# ax_f.set_title(r"Frecuencia: Dependencia recproca", fontsize=16, y=1.05)

arrow_f = ax_f.annotate('', xy=(0,0), xytext=(0,0), 
                        arrowprops=dict(arrowstyle='->', color='black', lw=1.5)) # Flecha simple apuntando al nulo
label_f = ax_f.text(0, 0, '', fontsize=14, ha='center', fontweight='bold')

fig.text(0.53, 0.5, r'$\overset{\mathcal{F}}{\longleftrightarrow}$', fontsize=30, ha='center', va='center')
# fig.suptitle("Reciprocidad: Tringulo vs Sinc", fontsize=20, fontweight='bold')

# ==========================================
# ANIMACIN
# ==========================================
def update(tau):
    # 1. Datos
    y_t = triangular_pulse(t, tau)
    y_f = sinc_squared_freq(w, tau)
    
    line_t.set_data(t, y_t)
    line_f.set_data(w, y_f)
    
    h_arrow = 0.1
    
    # --- ACTUALIZAR TIEMPO ---
    # Ahora la flecha es vertical y apunta al lmite del tringulo (tau)
    
    # La PUNTA (xy) va al eje cero en la posicin tau
    # 1. La PUNTA de la flecha (xy) debe ir SIEMPRE al valor exacto
    #    (mientras est dentro o en el borde del grfico)
    tip_pos = min(tau, LIMITE_T) 
    arrow_t.xy = (tip_pos, 0)
    
    # 2. La COLA/TEXTO (xytext) es lo que frenamos un poco antes 
    #    para que el texto no se corte al salirse del margen derecho.
    #    Dejamos un margen de 0.5 unidades respecto al borde.
    text_pos_x = min(tau, LIMITE_T * 0.9)
    text_pos_x = max(text_pos_x, tau * 0.65)

    # Si el nulo real est fuera de la grfica, levantamos un poco la flecha
    # o cambiamos el estilo, pero aqu simplemente inclinamos la flecha
    # si estamos muy cerca del borde.
    arrow_t.set_position((text_pos_x, 0.5))
    label_t.set_position((text_pos_x, 0.55))

    label_t.set_text(rf"$\tau = {tau:.2f}$")
    
    # --- ACTUALIZAR FRECUENCIA (CORRECCIN AQU) ---
    # Posicin matemtica EXACTA del nulo
    w_null_exact = 2 * np.pi / tau
    
    # 1. La PUNTA de la flecha (xy) debe ir SIEMPRE al valor exacto
    #    (mientras est dentro o en el borde del grfico)
    tip_pos = min(w_null_exact, LIMITE_W) 
    arrow_f.xy = (tip_pos, 0)
    
    # 2. La COLA/TEXTO (xytext) es lo que frenamos un poco antes 
    #    para que el texto no se corte al salirse del margen derecho.
    #    Dejamos un margen de 0.5 unidades respecto al borde.
    text_pos_x = min(w_null_exact, LIMITE_W * 0.9)
    text_pos_x = max(text_pos_x, LIMITE_W * 0.3)

    # Si el nulo real est fuera de la grfica, levantamos un poco la flecha
    # o cambiamos el estilo, pero aqu simplemente inclinamos la flecha
    # si estamos muy cerca del borde.
    arrow_f.set_position((text_pos_x, 0.5))
    label_f.set_position((text_pos_x, 0.55))
    
    label_f.set_text(rf"$\frac{{2\pi}}{{\tau}} = {w_null_exact:.2f}$")

# ==========================================
# GENERAR
# ==========================================
# Variacin geomtrica
taus = np.geomspace(1/LIMITE_T, LIMITE_T, 70)
taus = np.concatenate([taus, taus[::-1]])

print("Generando animacin con frmula explicita...")
anim = FuncAnimation(fig, update, frames=taus, interval=50, blit=False)
anim.save('reciprocidad_TF_tri.gif', writer=PillowWriter(fps=20), dpi=100)
print("Hecho!")

plt.show()
```
