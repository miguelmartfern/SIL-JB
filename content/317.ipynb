### Propiedades y pares de la SF en tiempo continuo

En la {ref}`tab:T3:SF:propiedades` se recogen las principales propiedades de la serie de Fourier para señales periódicas de período fundamental $T_0$ y potencia finita, mientras que en la {ref}`tab:T3:SF:pares` se muestran los pares transformados más comunes.

:::{table} Propiedades de la serie de Fourier
:name: tab:T3:SF:propiedades

| Propiedad | Señal periódica | Coef. Serie de Fourier |
| :--- | :--- | :--- |
|  | $\left.\begin{array}{c}\tilde{x}(t)\\ \tilde{y}(t)\end{array}\right\}$ Período $T_0$ <br> ($\omega_0=\frac{2\pi}{T_0}$) | $\begin{array}{c}c_k\\ d_k\end{array}$ |
| **Linealidad** | $A\tilde{x}(t)+B\tilde{y}(t)$ | $A\,c_k+B\,d_k$ |
| **Desplazamiento temporal** | $\tilde{x}(t-t_0)$ | $c_k\,e^{-jk\omega_0 t_0}$ |
| **Desplazamiento en frecuencia** | $e^{jM\omega_0 t}\tilde{x}(t)$ | $c_{k-M}$ |
| **Inversión temporal** | $\tilde{x}(-t)$ | $c_{-k}$ |
| **Conjugación** | $\tilde{x}^*(t)$ | $c^*_{-k}$ |
| **Escalado temporal** | $\tilde{x}(\alpha t)$, $\alpha>0$ <br> *(Periódica con período $T_0/\alpha$)* | $c_k$ |
| **Convolución periódica** | $\tilde{x}(t)\circledast \tilde{y}(t)$ | $T_0\,c_k d_k$ |
| **Multiplicación** | $\tilde{x}(t)\tilde{y}(t)$ | $c_k * d_k$ |
| **Diferenciación** | $\dfrac{d}{dt}\tilde{x}(t)$ | $jk\omega_0\,c_k$ |
| **Integración** | $\displaystyle \int_{-\infty}^{t}\tilde{x}(\tau)\,d\tau$ | $\left(\dfrac{1}{jk\omega_0}\right)c_k$ <br> *(Finita y periódica sólo si $c_0=0$)* |
:::

**Relación de Parseval:**
La potencia media de la señal periódica equivale a la suma de las potencias de sus armónicos:

```{math}
:enumerated: false
	\frac{1}{T_0}\int_{T_0}|\tilde{x}(t)|^2dt = \sum_{k=-\infty}^{+\infty}|c_k|^2
```

```{note} Sobre la relación de Parseval

Esta relación es consecuencia de que las exponenciales complejas relacionadas armónicamente forman una base ortogonal del espacio de las señales periódicas de periodo $T_0$.

Se puede ver como el teorema de Pitágoras aplicado a espacios de señales de dimensión infinita: mediante la serie de Fourier descomponemos la señal en componentes perpendiculares entre sí, formando una base.
```

:::{table} Pares Básicos de Series de Fourier
:name: tab:T3:SF:pares

| Señal periódica $\tilde{x}(t)$ | Coef. Serie de Fourier $c_k$ |
| :--- | :--- |
| $\displaystyle \sum_{k=-\infty}^{+\infty}c_k e^{jk\omega_0t}$ | $c_k$ |
| $e^{j\omega_0t}$ | $c_1=1$<br>$c_k=0,\quad k\neq 1$ |
| $\cos \omega_0 t$ | $c_1=c_{-1}=\frac{1}{2}$<br>$c_k=0, \quad k\neq \pm 1$ |
| $\sin \omega_0 t$ | $c_1=-c_{-1}=\frac{1}{2j}$<br>$c_k=0, \quad k\neq \pm 1$ |
| $1$ | $c_0=1$<br>$c_k=0, \quad k\neq 0$ |
| **Onda cuadrada periódica**<br>$\tilde{x}(t)=\begin{cases}1,&\|t\|<T_1\\0,& T_1<\|t\|\leq \frac{T}{2}\end{cases}$<br>$\tilde{x}(t+T)=\tilde{x}(t)$ | <br> $\displaystyle \frac{\sin(k\omega_0 T_1)}{k\pi} = \frac{\omega_0 T_1}{\pi}\text{sinc}\left(\frac{k\omega_0T_1}{\pi}\right)$ |
| **Tren de deltas**<br>$\displaystyle \sum_{n=-\infty}^{+\infty}\delta(t - nT)$ | <br> $\displaystyle c_k=\frac{1}{T},\quad \forall k$ |
:::

A continuación veremos con más detalle cada una de las propiedades y un ejemplo de aplicación de cada una de ellas.

#### Linealidad

:::{important .simple icon=false} Linealidad
Consideramos dos señales $\tilde{x}(t)$ e $\tilde{y}(t)$ periódicas con el mismo periodo $T_0$[^foot1].
```{math}
:enumerated: false
    \begin{split}
        \tilde{x}(t) \SF c_k\\
        \tilde{y}(t) \SF d_k
    \end{split}
```

[^foot1]: Recordemos que si las dos señales periódicas no tienen el mismo periodo fundamental, para que la señal $\tilde{z}(t)$ sea periódica, ambas señales tienen que tener un periodo en común, que será el mínimo común múltiplo del periodo fundamental de cada una de ellas:
    ```{math}
        T_0 = M T_{0x} = N T_{0y}, \quad M, N \in \mathbb{Z}.
    ```
    Además, habrá que hacer el desarrollo en serie de Fourier de ambas señales con el periodo común $T_0$.


```{math}
    \boxed{\tilde{z}(t)=A\tilde{x}(t)+B\tilde{y}(t) \SF a_k=A c_k + B d_k}
```
:::

```{math}
    \end{split}
```
```{math}
    \tilde{z}(t) =3\cos(2\pi t)-5j\sin(2\pi t) \SF ??
```

:::{solution} ejemplo:T3:SF:propiedades:linealidad:ejemplo2
:class: dropdown

Como el desarrollo en serie de Fourier de $\tilde{x}(t)$ e $\tilde{y}(t)$ está hecho con distinto periodo, **no podemos aplicar la propiedad de linealidad**.

Para poder aplicar la propiedad de linealidad, en primer lugar debemos buscar si ambas señales tienen un periodo en común[^foot1]:

```{math}
T_0=M T_{0x} = N T_{0y}, \, N, M \in \mathbb{Z} \Rightarrow T_0 = 1\, (M=1, N=2).
```

Obtenemos el desarrollo en serie de Fourier para $\tilde{y}(t)$ con periodo $T_0=2$ identificando con la ecuación de síntesis:
```{math}
    \tilde{y}(t)=\sum_{k=-\infty}^\infty d_k e^{jk2\pi t} \quad \Rightarrow \quad
    d_k=\begin{cases}
        \frac{-1}{2j}, & k=-2,\\
        \frac{1}{2j}, & k=2,\\
        0, & k\neq \pm 2.
    \end{cases}
```

```{math}
    \begin{split}
        \tilde{z}(t) &=3\cos(2\pi t)-5j\sin(2\pi t) \SF a_k=3c_k-5j d_k \\
        & \omega_0=2\pi, T_0=1.
    \end{split}
```

```{math}
        a_k=\begin{cases}
            \frac{5}{2}, & k=-2,\\
            \frac{3}{2}, & k=-1,\\
            \frac{3}{2}, & k=1,\\
            -\frac{5}{2}, & k=2,\\
            0, & k\neq \{\pm 1, \pm 2\}.
        \end{cases}
```
:::

#### Desplazamiento en el tiempo

:::{important .simple icon=false} Desplazamiento en el tiempo
Consideramos dos señales $\tilde{x}(t)$ de periodo $T_0$.
```{math}
:enumerated: false
    \tilde{x}(t) \SF c_k\\
```
```{math}
    \boxed{\tilde{y}(t)=\tilde{x}(t-t_0) \SF d_k=c_k e^{-jk\omega_0 t_0}}
```
:::

:::{exercise} Ejemplo 3
:label: ejemplo:T3:SF:propiedades:desp_t:ejemplo1

```{math}
    \begin{split}
        \tilde{x}(t) &=\cos(2\pi t) \SF \quad 
        c_k=\begin{cases}
            \frac{1}{2}, & k=\pm 1,\\
            0, & k\neq \pm 1.
        \end{cases}\\
        & \omega_0=2\pi, T_0=1.
    \end{split}
```
```{math}
    \tilde{y}(t)=\cos\left[2\pi\left(t-\frac{1}{4}\right)\right]=\tilde{x}\left(t-\frac{1}{4}\right) \SF ??
```

:::

:::{solution} ejemplo:T3:SF:propiedades:desp_t:ejemplo1
:class: dropdown

```{math}
    \begin{split}
        \tilde{y}(t) &=\cos\left[2\pi\left(t-\frac{1}{4}\right)\right]=\tilde{x}\left(t-\frac{1}{4}\right) \SF d_k = c_k e^{-jk2\pi\left(\frac{1}{4}\right)} \\
        & \omega_0=2\pi, T_0=1.
    \end{split}
```

```{math}
        d_k=\begin{cases}
            \frac{1}{2} e^{-j(-1) 2\pi \left(\frac{1}{4}\right)}=\frac{1}{2} e^{j\pi/2}=\frac{j}{2}, & k=-1,\\
            \frac{1}{2} e^{-j(1) 2\pi \left(\frac{1}{4}\right)}=\frac{1}{2} e^{-j\pi/2}=\frac{-j}{2}, & k=1,\\
            0, & k\neq \pm 1.
        \end{cases}
```

Vamos a comprobarlo identificando con la ecuación de síntesis directamente:
```{math}
    \begin{split}
        \tilde{y}(t) &=\cos\left[2\pi\left(t-\frac{1}{4}\right)\right]=\frac{e^{j2\pi\left(t-\frac{1}{4}\right)}+e^{-j2\pi\left(t-\frac{1}{4}\right)}}{2}= \sum_{k=-\infty}^\infty d_k e^{jk2\pi t}\\
            &=\frac{1}{2}e^{-j\frac{\pi}{2}}e^{j2\pi t}+\frac{1}{2}e^{j\frac{\pi}{2}}e^{-j2\pi t} = \frac{j}{2}e^{-j2\pi t}-\frac{j}{2}e^{j2\pi t}.
    \end{split}
```
:::

#### Desplazamiento en frecuencia

:::{important .simple icon=false} Desplazamiento en frecuencia
Consideramos dos señales $\tilde{x}(t)$ de periodo $T_0$.
```{math}
:enumerated: false
    \tilde{x}(t) \SF c_k\\
```

```{math}
    \tilde{y}(t)=e^{j6\pi t}\cos(2\pi t) \SF ??
```

:::

:::{solution} ejemplo:T3:SF:propiedades:desp_k:ejemplo1
:class: dropdown

```{math}
    \begin{split}
        \tilde{y}(t) &=e^{j6\pi t}\cos(2\pi t) \SF d_k=c_{k-3}  \\
        & \omega_0=2\pi, T_0=1.
    \end{split}
```

```{math}
        d_k=\begin{cases}
            \frac{1}{2}, & k=2,\\
            \frac{1}{2}, & k=4,\\
            0, & k\neq \{2,4\}.
        \end{cases}
```

Vamos a comprobarlo identificando con la ecuación de síntesis directamente:
```{math}
    \begin{split}
        \tilde{y}(t) &=e^{j6\pi t}\cos(2\pi t)=e^{j6\pi t}\frac{e^{j2\pi t}+e^{j2\pi t}}{2}= \sum_{k=-\infty}^\infty d_k e^{jk2\pi t}\\
        &=\frac{1}{2}e^{j8\pi t}+\frac{1}{2}e^{j4\pi t}.
    \end{split}
```
